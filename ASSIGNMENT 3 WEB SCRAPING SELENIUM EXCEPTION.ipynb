# QUESTION 2
In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a data frame and csv. In case if any product has less than 3 pages
in search results then scrape all the products available under that product name. Details to be scraped are:
    "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
start_page = 0
end_page = 2
urls = []
for page in range(start_page,end_page+1):
    try:
        page_urls = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')
        for url in page_urls:
            url = url.get_attribute('href')     
            if url[0:4]=='http':                
                urls.append(url)                
        print("Product urls of page {} has been scraped.".format(page+1))
        
        next_page=driver.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").get_attribute('href')
        driver.get(next_page)
        driver.refresh()   
    except:
        break
```

```python
len(urls)
```

```python
brand_list=[]
title_list=[]
price_list=[]
return_list=[]
availability_list=[]
delivery_list=[]
img_url_list=[]
     
for x in urls:
    driver.get(x)
    
    try:
        brand=driver.find_element(By.ID,"productTitle")
        brand_list.append(brand.text)
    except NoSuchElementException as e:
        brand_list.append("--")
    
    try:
        title=driver.find_element(By.XPATH,"//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
        title_list.append(title.text)
    except NoSuchElementException as e:
        title_list.append("--")
    
    try:
        price=driver.find_element(By.XPATH,"//span[@class='a-price-whole']")
        price_list.append(price.text)
    except NoSuchElementException as e:
        price_list.append("--")  
  
    try:
        returns=driver.find_element(By.XPATH,"//a[@class='a-size-small a-link-normal a-text-normal']")
        return_list.append(returns.text)
    except NoSuchElementException as e:
        return_list.append("--")
    
    try:
        availability=driver.find_element(By.XPATH,"//span[@class='a-size-medium a-color-success']")
        availability_list.append(availability.text)
    except NoSuchElementException as e:
        availability_list.append("--")   
     
    try:
        delivery=driver.find_element(By.XPATH,"//span[@class='a-color-base a-text-bold']")
        delivery_list.append(delivery.text.replace("\n"," "))
    except NoSuchElementException as e:
        delivery_list.append("--")
     
    try:
        img_url=driver.find_element(By.XPATH,"//div[@class='imgTagWrapper']/img")
        img_url_list.append(img_url.get_attribute("src"))
    except NoSuchElementException as e:
        img_url_list.append("--")
```

```python
print(len(brand_list))
print(len(title_list))
print(len(price_list))
print(len(return_list))
print(len(availability_list))
print(len(delivery_list))
print(len(img_url_list))
```

```python
df=pd.DataFrame({})
df['Brand']=brand_list
df['Name of the product']=title_list
df['Price']=price_list
df['Return/Exchange']=return_list
df['Availability']=availability_list
df['Delivery']=delivery_list
df['Product Urls']=img_url_list
df.style.set_properties(**{'text-align':'left'}).set_table_styles([ dict(selector='th', props=[('text-align', 'left')])])
```
